# Configuration

```python
class config:
    data="data/"
    configs=["sift", "aliked2k", "disk"]
    retrieval="dinov2_salad"
    n_retrieval= 25
    n_retrieval_transp = 30
    mode="train"
    output="./temp"
    seed=0
    hloc_script_path="./scripts/run_hloc.py"
    pixsfm=True
    pixsfm_force=False
    pixsfm_script_path="./scripts/run_pixsfm.py"
    pixsfm_max_imgs= 9999
    pixsfm_config_ref="low_memory.yaml"
    pixsfm_config="config/pixsfm.yaml"
    rotation_matching=True
    rotation_wrapper=False
    resize=None
    equalize_transp=True
    equalize_grayscale=False
    equalize_grid_size=8
    equalize_clip_limit=60.
    shared_camera=True
    overwrite=True
    cropping=False
    max_rel_crop_size=0.75
    min_rel_crop_size=0.2
    #skip_scenes=None
    skip_scenes=["lizard", "dioscuri", "transp_obj_glass_cylinder", "transp_obj_glass_cup", "multi-temporal-temple-baalshamin", "pond"]
    localize_unregistered=True
```

# Run Pipeline

```python
# Adjust pixsfm config
import yaml
from pathlib import Path
import pixsfm.configs

conf_path = Path(pixsfm.configs.__file__).parent / config.pixsfm_config_ref

with open(conf_path, 'r') as f:
    conf = yaml.load(f, Loader=yaml.FullLoader)

if config.pixsfm_config_ref == "low_memory.yaml":
    conf["mapping"]["BA"]["optimizer"]["refine_focal_length"] = True
    conf["mapping"]["BA"]["optimizer"]["refine_extra_params"] = True
    conf["mapping"]["BA"]["optimizer"]["refine_extrinsics"] = True

    # Remove randomness
    conf["mapping"]["BA"]["costmaps"]["num_threads"] = 1
    conf["mapping"]["BA"]["num_threads"] = 1
    conf["mapping"]["KA"]["optimizer"]["num_threads"] = 1

else:
    conf["dense_features"]["use_cache"] = True
    conf["dense_features"]["overwrite_cache"] = True

    #conf["mapping"]["dense_features"] = "${..dense_features}"
    conf["mapping"]["BA"]["optimizer"]["refine_focal_length"] = True
    conf["mapping"]["BA"]["optimizer"]["refine_extra_params"] = True
    conf["mapping"]["BA"]["optimizer"]["refine_extrinsics"] = True
    
    # Remove randomness
    conf["mapping"]["BA"]["optimizer"]["solver"]["num_threads"] = 1
    conf["mapping"]["BA"]["optimizer"]["num_threads"] = 1
    conf["mapping"]["BA"]["references"]["num_threads"] = 1
    conf["mapping"]["BA"]["num_threads"] = 1

    conf["mapping"]["KA"]["optimizer"]["num_threads"] = 1
    conf["mapping"]["KA"]["optimizer"]["solver"]["num_threads"] = 1

    conf["localization"]["QBA"]["optimizer"]["solver"]["num_threads"] = 1
    conf["localization"]["QKA"]["optimizer"]["solver"]["num_threads"] = 1
    conf["localization"]["references"]["num_threads"] = 1
    #conf["localization"]["dense_features"] = "${..dense_features}"

out_conf_path = Path("config", "pixsfm.yaml")
out_conf_path.parent.mkdir(exist_ok=True, parents=True)
with open(out_conf_path, 'w') as f:
    yaml.dump(conf, f)
```

```python
import argparse
from pathlib import Path
from pixhloc.main import main

args = {
    "data": config.data,
    "configs": config.configs,
    "retrieval": config.retrieval,
    "n_retrieval": config.n_retrieval,
    "n_retrieval_transp": config.n_retrieval_transp,
    "mode": config.mode,
    "output": config.output,
    "seed": config.seed,
    "hloc_script_path": config.hloc_script_path,
    "pixsfm": config.pixsfm,
    "pixsfm_script_path": config.pixsfm_script_path,
    "pixsfm_force": config.pixsfm_force,
    "pixsfm_max_imgs": config.pixsfm_max_imgs,
    "pixsfm_config": config.pixsfm_config,
    "rotation_matching": config.rotation_matching,
    "rotation_wrapper": config.rotation_wrapper,
    "resize": config.resize,
    "equalize_transp": config.equalize_transp,
    "equalize_grayscale": config.equalize_grayscale,
    "equalize_grid_size": config.equalize_grid_size,
    "equalize_clip_limit": config.equalize_clip_limit,
    "shared_camera": config.shared_camera,
    "overwrite": config.overwrite,
    "cropping": config.cropping,
    "max_rel_crop_size": config.max_rel_crop_size,
    "min_rel_crop_size": config.min_rel_crop_size,
    "skip_scenes": config.skip_scenes,
    "localize_unregistered": config.localize_unregistered
}

args = argparse.Namespace(**args)
Path(args.output).mkdir(exist_ok=True)

main(args)
```
